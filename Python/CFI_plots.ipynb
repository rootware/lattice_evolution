```python
qfi_data = np.loadtxt("../testing_QFI_basis/unchanged_momentum_basis.txt");
```

```python
def I_MZ( time):
    return (4*4* (time/2)**2)**2;
```

```python
f= lambda x: x**4;
plt.plot(qfi_data[:,0]/Units.omega_R*1/10**(-6), qfi_data[:,1]/f(qfi_data[:,0]), label="cfi")
plt.plot(qfi_data[:,0]/Units.omega_R*1/10**(-6), qfi_data[:,2]/f(qfi_data[:,0]), label="qfi")
```

```python
plt.plot(qfi_data[:,0]/Units.omega_R*1/10**(-6), qfi_data[:,1]/I_MZ(qfi_data[:,0]), label="$I_{aa}$")
plt.plot(qfi_data[:,0]/Units.omega_R*1/10**(-6), qfi_data[:,2]/I_MZ(qfi_data[:,0]), label="$F_{aa}$")


plt.ylabel("$I_{aa}(t)$ and $F_{aa}(t)$ in $I_{MZ}(t)$")
plt.xlabel("$t(\mu s)$")
plt.legend()
plt.title("CFI and QFI in $I_{MZ}(t)$ units")
```

```python
plt.plot(qfi_data[:,0]/Units.omega_R*1/10**(-6), qfi_data[:,3]/I_MZ(qfi_data[:,0]), label="$I_{VV}$")
plt.plot(qfi_data[:,0]/Units.omega_R*1/10**(-6), qfi_data[:,4]/I_MZ(qfi_data[:,0]), label="$F_{VV}$")

plt.ylabel("$I_{VV}(t)$ and $F_{VV}(t)$ in $I_{MZ}(t)$")
plt.xlabel("$t(\mu s)$")
plt.legend()
plt.title("CFI and QFI in $I_{MZ}(t)$ units")
```

```python
plt.loglog(qfi_data[:,0]/Units.omega_R*1/10**(-6), qfi_data[:,3]/I_MZ(qfi_data[:,0]), label="$I_{VV}$")
plt.loglog(qfi_data[:,0]/Units.omega_R*1/10**(-6), qfi_data[:,4]/I_MZ(qfi_data[:,0]), label="$F_{VV}$")

plt.ylabel("$\log I_{VV}(t)$ and $\log F_{VV}(t)$ in $I_{MZ}(t)$")
plt.xlabel("$\log t(\mu s)$")
plt.legend()
plt.title("$\log I$ and $\log F$ in $I_{MZ}(t)$ units")
```

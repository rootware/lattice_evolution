```python
cfi_data = np.loadtxt("../Recom0/cfi_qfi_longer_withprop.txt")
cfi_data  = cfi_data[ np.lexsort((cfi_data[:,1],cfi_data[:,0])) ]
```

```python
import seaborn as sns;
import seaborn.objects as so;
sns.set_theme()
```

```python
shakingfunction = np.loadtxt("../Recom0/combined_shaking.txt")
time = np.arange(0, len(shakingfunction))*50e-9*Units.omega_R
plt.plot(time,shakingfunction)
plt.ylabel("$\phi(t)$")
plt.xlabel("$t(\omega_R^{-1})$")
plt.title("Recom0")
plt.savefig("Recom0_shaking.pdf")
```

```python

def I_MZ( time):
    return (4*4* (time/2)**2)**2;

plt.plot(AList*Units.accUnit/Units.g, cfi_data[:,2]/I_MZ(time[-1]), label="CFI")
plt.plot(AList*Units.accUnit/Units.g,cfi_data[:,3]/I_MZ(time[-1]), label="QFI")
plt.title("Fisher Information Normalized By MZ")
plt.xlabel("$a(g)$")
plt.legend()
plt.ylabel("Normalized QFI/CFI")
plt.savefig("Recom0_QFInCFI.pdf")
```
